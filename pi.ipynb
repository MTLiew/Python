```python
import math
import random
import matplotlib.pyplot as plt

points = []
inCircleCount = 0
inSquareCount = 0
pies = []
errors = []
squareplotpoints = []
circleplotpoints = []

for k in range(3, 7):
    for n in range(10 ** k):
        x = random.random()
        y = random.random()
        points.append((x, y))

        d = math.sqrt((x ** 2) + (y ** 2))

        if d <= 1:
            inCircleCount += 1
            inSquareCount += 1

            if k == 4:
                circleplotpoints.append((x, y))

        else:
            inSquareCount += 1

            if k == 4:
                squareplotpoints.append((x, y))

    myPi = 4 * (inCircleCount/inSquareCount)
    pies.append(myPi)

    error = abs((math.pi - myPi)/math.pi) * 100
    errors.append(error)

for k in range(3, 7):
    print("n = 10 ^", k, " pi =", "%07.6f"%pies[k - 3], " error =", "%05.4f"%errors[k - 3], "%")

xc, yc = zip(*circleplotpoints)
xs, yx = zip(*squareplotpoints)

plt.scatter(*zip(*circleplotpoints), s=10, color="blue")
plt.scatter(*zip(*squareplotpoints), s=10, color="red")
plt.show()
```
